```python
import numpy as np
import matplotlib.pyplot as plt

N = 100000
a = 0
b = np.pi
x = np.random.uniform(a,b,N) #N es el número de muestras aleatorias que tomará
```

```python
def func_integrate(x):
    return np.exp(-x)*np.sin(x)
```

```python
fi = func_integrate(x)
```

```python
I = (b-a)*sum(fi)/N
print(I)
```

```python
Iteo = 0.5*(1+np.exp(-np.pi))
print(f'El valor teórico es: {Iteo:.6f} y valor utilizando Monte Carlo es {I:.6f}, con un error porcentual de: {np.abs(1-I/Iteo): .6%}')
```

```python
error_porcentual= []
N = 100 #se tomo este valor ya que con más demoraba mucho
a = 0
b = np.pi
x = np.random.uniform(a,b,N)

for N in x:
    fi = func_integrate(x)
    I = (b-a)*sum(fi)/N  #Aqui se calcula la integral con Montecarlo
    error = abs((I - Iteo) / Iteo) * 100
    error_porcentual.append(error)
```

```python

plt.figure(figsize=(12, 7))
plt.scatter(x, error_porcentual, color="r")
plt.title("Error Porcentual Integral vs. Número de Muestras N")
plt.xlabel("Número de Muestras N")
plt.ylabel("Error Porcentual (%)")
plt.grid(True)
plt.show()
```

```python
# utilizando escala logaritmica:

plt.figure(figsize=(12, 7))
plt.scatter(x, error_porcentual, color="r")
plt.title("Error Porcentual Integral vs. Número de Muestras N (escala Logaritmica)")
plt.xlabel("Número de Muestras N")
plt.ylabel("Error Porcentual (%)")
plt.xscale("log")
plt.yscale("log")
plt.grid(True)
plt.show()
```
